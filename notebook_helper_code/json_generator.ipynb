```python
title_abbrev="new_book"
title="NewBook"
stars="3"
day_view="2026/01/09"
author="Me"
review="This sucked"
media_type="book"
```

```python
if media_type=="movie":
    return_string=f'"{title_abbrev}": {{"title": "{title}", "stars": "{stars}", "day_watched": "{day_view}", "director": "{author}", "review": "{review}"}}'
    print(return_string+",")
elif media_type=="book":
    return_string=f'"{title_abbrev}": {{"title": "{title}", "stars": "{stars}", "day_read": "{day_view}", "author": "{author}", "review": "{review}"}}'

    print(return_string+",")
```
