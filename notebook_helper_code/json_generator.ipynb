```python
title_abbrev="bone_temp"
title="28 Years Later: The Bone Temple"
stars="4"
day_view="2026/01/16"
author="Nia DeCosta"
review="Give me the 3 hour Samson/Fiennes hangout movie."
media_type="movie"
```

```python
if media_type=="movie":
    return_string=f'"{title_abbrev}": {{"title": "{title}", "stars": "{stars}", "day_watched": "{day_view}", "director": "{author}", "review": "{review}"}}'
    return_string=return_string+","
elif media_type=="book":
    return_string=f'"{title_abbrev}": {{"title": "{title}", "stars": "{stars}", "day_read": "{day_view}", "author": "{author}", "review": "{review}"}}'
    return_string=str(repr(return_string+","))
```

```python
print(return_string)
```

```python
#books need a solution to the /n problem before it can be used here
file="movies.json"
if media_type=="book": 
    file="books.json"
    
import os 
current_directory = os.getcwd()
curr_file=os.path.basename(current_directory)
directory=current_directory.split(curr_file)[0]
with open(f'{directory}/{file}', 'r') as f: 
    content=f.read()
    new_content=content[0]+return_string+content[1:]
with open(f'{directory}/{file}',"w") as f: 
    f.write(new_content)
    
```
