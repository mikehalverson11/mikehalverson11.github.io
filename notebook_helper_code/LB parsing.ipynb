```python
import pandas as pd 
```

```python
df=pd.read_csv("/Users/michaelhalverson/Downloads/diary.csv")
```

```python
df=df[df["Watched Date"]>'2023-01-01']
df=df[["Watched Date","Name", "Rating"]]
df=df.rename(columns={"Watched Date": "day_watched", "Name": "title", "Rating": "stars"})
```

```python
json_title=df["title"].str.split(" ",expand=True)[0]+"_"+df["title"].str.split(" ",expand=True)[1]
json_title=json_title.str.replace(":","")
```

```python
df["json_title"]=json_title
df.loc[df.json_title.isnull(),"json_title"]=df[df.json_title.isnull()]["title"]
df["json_title"]=df["json_title"].str.lower()
```

```python
df["review"]=""
df["director"]=""
```

```python
df["json_object"]='"'+df["json_title"]+'": {"title": "' + df["title"]+'", "director": "' + df["director"] + '", "stars": "' + df["stars"].astype("str")+'", "review": "'+df["review"].str.replace('"',"'").replace("\t","") + '","day_watched": "' + df["day_watched"]+ '"}'
```

```python
json_string="{" + ','.join(df["json_object"]) + "}"
```

```python
import json
json.loads(json_string)
```

```python
with open("movies.json",'w') as file: 
    file.write(json_string)
```
